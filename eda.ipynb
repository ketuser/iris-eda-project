#### Step 2. Fetching and Loading the Dataset  

````python
# Necessary libraries were imported:
import pandas as pd  # For data manipulation
import pandasql as psql # SQL Querying
import numpy as np   # For numerical operations
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns  # For enhanced visualizations
````

````python
# The Iris dataset was fetched from UCI (direct URL, no API needed) and loaded into a DataFrame:
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']  # Define headers
df = pd.read_csv(url, header=None, names=column_names)  # Load into DataFrame
````

````python
# A quick preview of the data was performed:
print(df.head())      # First 5 rows
print(df.info())      # Data types and non-null counts
print(df.describe())  # Basic statistics
````

#### Step 3. Data Cleaning and Basic Exploration

````python
# Duplicates and missing values were checked
print(f"Duplicates: {df.duplicated().sum()}") 
df = df.drop_duplicates() 

# The species column was converted to categorical type:
df['species'] = df['species'].astype('category')

# Basic exploration of nique species was performed:
print(df['species'].value_counts())  

# Outliers were visualized using boxplots:
sns.boxplot(data=df.drop('species', axis=1))  # Visualizing distributions
plt.title('Boxplots for Features')
plt.show()

# Note: Iris has minimal outliers; if any, we can clip or ignore for simplicity
````

#### Step 4. Visualizations and Insights

````python
# Relationships between features were visualized using pairplots:
sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Iris Features by Species')
plt.show()

# Feature correlations were visualized using a heatmap:
numeric_df = df.drop('species', axis=1)
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````

Key insights were derived:

  - Petal length and width strongly correlate (0.96), useful for classification.

  - Setosa species has smaller petals, making it easier to distinguish.

  - Versicolor and Virginica overlap more, requiring more advanced models for separation.

Pairplot shows clear clusters by species; Heatmap confirms high petal correlations.

##### Step 5. Adding SQL Querying for Practice

````python
# pandasql was imported to simulate SQL queries on the DataFrame:
from pandasql import sqldf
````

````python
# An example query was executed to calculate average measurements by species:
query = """
SELECT species, AVG(sepal_length) as avg_sepal_length, AVG(petal_length) as avg_petal_length
FROM df
GROUP BY species
"""
result = sqldf(query)
print(result)
````
